````python
import os
import openpyxl
from openpyxl.styles.colors import COLOR_INDEX

# Path to the root directory
root_dir = "/home/simon/Documents/data_science/Thesis/manual_download/"

# Helper function to check if a cell is yellow
def is_yellow(cell):
    fill = cell.fill
    if fill is not None and fill.fill_type is not None:
        color = fill.fgColor
        # Check for standard yellow (RGB: FFFF00)
        if color.type == "rgb" and color.rgb is not None:
            return color.rgb.upper() in ["FFFFFF00", "FFFF00"]
        # Check for indexed yellow (index 5 is yellow in Excel's color index)
        if color.type == "indexed" and color.indexed is not None:
            return color.indexed == 5
    return False

# Walk through all subfolders and files
for dirpath, _, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith(".xlsx"):
            file_path = os.path.join(dirpath, filename)
            wb = openpyxl.load_workbook(file_path, data_only=True)
            for ws in wb.worksheets:
                yellow_rows = []
                for row in ws.iter_rows():
                    if any(is_yellow(cell) for cell in row):
                        yellow_rows.append([cell.value for cell in row])
                if yellow_rows:
                    print(f"File: {file_path}, Sheet: {ws.title}")
                    for row in yellow_rows:
                        print(row)
````
