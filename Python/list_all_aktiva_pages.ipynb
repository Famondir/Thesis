```python
import os
import csv

folder_path = "/home/simon/Documents/data_science/Thesis/micha_gold_truth"
pdf_files = []

for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.lower().endswith('.pdf'):
            pdf_files.append(os.path.join(root, file))

csv_path = os.path.join(folder_path, "pdf_files_list.csv")
with open(csv_path, "w", newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["pdf_path"])
    for pdf in pdf_files:
        writer.writerow([pdf])
```

```python
from io import BytesIO
from reportlab.pdfgen import canvas
from pypdf import PdfReader, PdfWriter

pdf_writer = PdfWriter()
output_pdf_path = "/home/simon/Documents/data_science/Thesis/micha_gold_truth/combined_aktiva_pages.pdf"

for entry in activa.itertuples(index=False):
    pdf_path = entry.pdf_path
    page_num = entry.page_aktiva - 1
    try:
        reader = PdfReader(pdf_path.replace("/Downloads/", "/Documents/data_science/Thesis/"))
        page = reader.pages[page_num]
        media_box = page.mediabox
        width = float(media_box.width)
        height = float(media_box.height)

        # Create overlay PDF with text, matching original page size
        packet = BytesIO()
        can = canvas.Canvas(packet, pagesize=(width, height))
        can.setFont("Helvetica", 8)
        can.drawString(10, height - 20, f"{pdf_path} | page {entry.page_aktiva}")
        can.save()
        packet.seek(0)
        overlay_pdf = PdfReader(packet)
        overlay_page = overlay_pdf.pages[0]

        # Merge overlay onto original page
        page.merge_page(overlay_page)
        pdf_writer.add_page(page)
    except Exception as e:
        print(f"Error processing {pdf_path} page {page_num+1}: {e}")

with open(output_pdf_path, "wb") as out_f:
    pdf_writer.write(out_f)

print(f"Combined PDF saved to {output_pdf_path}")
```
